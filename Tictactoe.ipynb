```python
def tictactoe():
    board=[[' ',' ',' '],[' ',' ',' '],[' ',' ',' ']]
    move_count=0
    from random import randint
    x=randint(1,2)
    y=input('Please toss the coin to determine who will be player 1')
    print('The result is {}'.format(x))
    print(' You won the toss, please go first. You will be "X" ')
    while check(board):
        row=0
        col=0
        if move_count%2==0:
            print('Player one, please make your move [x] ')
            row=int(input('Please enter row ' ))
            col=int(input('Please enter column '))
            if board[row-1][col-1] not in ['X','O']:
                board[row-1][col-1]='X'
                move_count+=1
            else:
                print('Invalid move, try again')
                continue
            check(board)
        
                
        else:
            print('Player two, please make your move [o] ')
            row=int(input('Please enter row ' ))
            col=int(input('Please enter column '))
            if board[row-1][col-1] not in ['X','O']:
                board[row-1][col-1]='O'
                move_count+=1
            else:
                print('Invalid move, try again')
                continue
                
            check(board)
            
        success=check(board)
        print('1',board[0],'\n''2',board[1],'\n''3',board[2])
     
        if move_count==9 : break
    
    if success:
        print('Stalemate, no one wins')
    else:
        if move_count%2==0:
            print("Player 2 has won!!!!!!")
        else:
            print("Player 1 has won!!!!!!")
        
    return None
    
    
def check(x):
    if row_check(x[0]) or row_check(x[1]) or row_check(x[2]):
        return False
    elif col_check([x[0][0],x[1][0],x[2][0]]) or col_check([x[0][1],x[1][1],x[2][1]]) or col_check([x[0][2],x[1][2],x[2][2]]):
        return False
    elif diag_check(x[0][0],x[1][1],x[2][2]) or diag_check(x[0][2],x[1][1],x[2][0]):
        return False
        
    return True
        
    
def row_check(y=[0,0,0]):
    if y[0]==y[1]==y[2] and y[0]!= ' ':
        return True
    else:
        return False
    
def col_check(y=[0,0,0]):
    if y[0]==y[1]==y[2] and y[0]!= ' ':
        return True
    else:
        return False
    
def diag_check(a,b,c):
    if a==b==c and a!=' ':
        return True
    else:
        return False
   
```
